```python
import pandas as pd
import pyshark
import json
import collections
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
import re
from collections import defaultdict
```

```python
http = pyshark.FileCapture('Data/http.pcap', keep_packets=True)
http.load_packets()
print(len(http))

def extractPacketColumn(data):
    df_data = []
    ETH = 0
    IP = 1
    TCP = 2
    for pkt in data:
        if len(pkt.layers) < 3:
            continue
        if pkt[2].layer_name != 'tcp':
            continue
        
        d = {
            'timestamp': pkt.sniff_time.timestamp(),
            'length': pkt.length,
            'highest_layer': pkt.highest_layer,
            'dest_ip': pkt[IP].dst,
            'dest_port': pkt[TCP].dstport,
            'src_ip': pkt[IP].src,
            'src_port': pkt[TCP].srcport,
        }
        df_data.append(d)
    return pd.DataFrame(df_data)
df_http = extractPacketColumn(http)

dns = pyshark.FileCapture('Data/dns.pcap', keep_packets=True)
dns.load_packets()
print(len(dns))
df_dns = extractPacketColumn(dns)
```
